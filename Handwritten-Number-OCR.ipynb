```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import cv2

import random
from scipy.signal import find_peaks
```

```python
DATADIR = "D:/Programming/Roshan/ORAND-CAR-2014/ORAND-CAR-2014"
train_dir = [["CAR-A/a_train_gt.txt","CAR-A/a_train_images"], ["CAR-B/b_train_gt.txt", "CAR-B/b_train_images"]]

RESIZED_IMG_HEIGHT = 60
RESIZED_IMG_WIDTH = 200
RESIZED_DIGIT_SIZE = 60

def create_data(X, y, dirs):
    data_array = []
    
    for gt_dir, img_dir in dirs:
        print(gt_dir, "---", img_dir)
        gt = open(os.path.join(DATADIR, gt_dir))
        img_path = os.path.join(DATADIR, img_dir)

        for data in gt:
            img_name, img_ans = data.split()
            img_array = cv2.imread(os.path.join(img_path, img_name), cv2.IMREAD_GRAYSCALE)
            scaled_img_array = cv2.resize(img_array, (RESIZED_IMG_WIDTH, RESIZED_IMG_HEIGHT))
            data_array.append([scaled_img_array, img_ans])
            
    random.shuffle(data_array)
    
    for features, label in data_array:
        X.append(features)
        y.append(label)
    X = np.array(X).reshape(-1,  RESIZED_IMG_HEIGHT,RESIZED_IMG_WIDTH, 1)
    X = tf.keras.utils.normalize(X, axis=1)  
    
    return X, y
    

X_train = []
y_train = []
X_train, y_train  = create_data(X_train, y_train, train_dir)
```

```python
def slice_to_digits(X_digits, y_digits, X_full, y_full):
    ct=0
    broken_counter = 0
    digit_ct = 0
    
    for full_num_image, full_num_ans in zip(X_full, y_full):
        mean_array = np.mean(full_num_image, axis=0)
        peaks = find_peaks(mean_array.reshape(1, -1)[0], distance=20, height=0.126)[0]    # prominence=0.0001

        if len(peaks) > 0 and peaks[0] > 15:
            np.insert(peaks, 0, 1)
        if len(peaks) > 0 and peaks[-1] < 170:
            np.append(peaks, 200)
            
        image_parts_array = []
        for i in range(len(peaks)-1):
            start = peaks[i]
            end = peaks[i + 1]
            digit_image = full_num_image[:, start:end]
            if np.sum(digit_image) < 450:
                image_parts_array.append(cv2.resize(digit_image, (RESIZED_DIGIT_SIZE, RESIZED_DIGIT_SIZE)))
    

        if len(full_num_ans) == len(image_parts_array):
            ct += 1
            for i in range(len(full_num_ans)):
                digit_ct += 1
                X_digits.append(cv2.resize(image_parts_array[i], (RESIZED_DIGIT_SIZE, RESIZED_DIGIT_SIZE)))
                y_digits.append(int(full_num_ans[i]))
        else:
            broken_counter += 1
            
    print(f"Sliced : {ct}")
    print(f"UnSliced: {broken_counter}")
    print(f"Digits Counted: {digit_ct}")
    return X_digits, y_digits
    

X_train_digits = []
y_train_digits = []
X_train_digits, y_train_digits = slice_to_digits(X_train_digits, y_train_digits, X_train, y_train)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))

model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

model.fit(np.array(X_train_digits), np.array(y_train_digits), epochs=5)
```
